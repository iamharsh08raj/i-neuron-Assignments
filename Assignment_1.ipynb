2. What is the difference between string and variable?
A variable is where we store values and a string is a type of value stored in a variable.
Eg: x=’Harsh’
Here x is a variable and ‘Harsh’ is a string.


3. Describe three different data types.
Integer- Integer is a data type which stores whole numbers.
Eg- 1,2,3,etc.
String- String is a data type that stores sequence of characters.
Eg- “Harsh”
Boolean- String is a data type that stores two values True & False.


4. What is an expression made up of? What do all expressions do?
An expression consists of operators and operands that is interpreted to produce some other value. An expression is evaluated as per the precedence of its operators.
